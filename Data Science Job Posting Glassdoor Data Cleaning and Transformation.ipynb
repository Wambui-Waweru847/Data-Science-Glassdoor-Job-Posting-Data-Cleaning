#### Data Science Job Posting Glassdoor Data Cleaning and Transformation

```python
#import libraries
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Uncleaned_DS_jobs.csv", index_col= 0)
```

```python
#Check teh overview of the dataset
df.shape
df.info()
df.head()
```

```python
#Drop unnecessary columns
df.drop("Competitors", inplace=True, axis=1)
```

```python
#Fixing the job title column
df["Job Title"].value_counts()
```

```python
df["Job Title"].unique()
```

```python
#Categorise the job titles while reducing them
def job_listing(title):
    if "data scientist" in title.lower():
        return "Data Scientist"
    elif "data engineer" in title.lower():
        return "Data Engineer"
    elif "analyst" in title.lower():
        return "Analyst"
    elif "manager" in title.lower():
        return "Manager"
    elif "director" in title.lower():
        return "Director"
    else:
        "Other"
```

```python
df["job_title"] = df["Job Title"].apply(job_listing)
df["job_title"].value_counts()
```

```python
#Assigning job titles by their levels
def job_title_level(title):
    if "sr" in title.lower() or "senior" in title.lower() or "principal" in title.lower() or "lead" in title.lower():
        return "Senior"
    elif "jr" in title.lower() or "jr." in title.lower():
        return "Junior"
    else:
        return "Other"
```

```python
df["job_level"] = df["Job Title"].apply(job_title_level)
df["job_level"].value_counts()
```

```python
df.drop(['Job Title'], axis = 1, inplace = True)
df.head()
```

```python
#Extract the company name only since the column has mixed values
df["company_name"] = df["Company Name"].map(lambda x: x.split("\n")[0])
df["company_name"].head()
df.drop("Company Name", inplace=True, axis=1)
```

```python
df.tail(10)
```

```python
#Extract the min and max salary columns
pattern = r'\$(\d+)K-\$(\d+)K'
df[["min_salary", "max_salary"]]= df["Salary Estimate"].str.extract(pattern).astype(int)
#Calculating avg salary
df["avg_salary"] = df.apply(lambda row: int((row["min_salary"] + row["max_salary"])/2), axis=1)
df["Salary Estimate"] = df["min_salary"].astype(str) + "-" + df["max_salary"].astype(str)
df.drop("Salary Estimate", inplace=True, axis=1)
```

```python
df.head()
```

```python
#Handling inconsistent data
# list all the categorical columns
categorical_cols = ["Job Description",  "Location", "Headquarters", "Size", "Type of ownership", "Industry", "Sector", "Revenue", ]
#Replace -1 and "unknown" with "na"
df[categorical_cols] = df[categorical_cols].replace({"-1" :"na", "Unknown": "na"})

# Replace -1.0 with 0 in the rating column
df["Rating"] = df["Rating"].replace(-1.0, 0)

df.head()
```

```python
#Check the job description column
df["Job Description"][0].split('\n')
```

```python
#Categorising skills based popular technologies used in Data Science.
def machine_learning(skill):
    if "machine learning" in skill.lower():
        return 1
    else:
        return 0
df["Machine Learning"] = df["Job Description"].apply(machine_learning)
    
def big_data(skill):
    if "big data" in skill.lower():
        return 1
    else:
        return 0
df["Big Data"] = df["Job Description"].apply(big_data)
def excel(skill):
    if "excel" in skill.lower():
        return 1
    else:
        return 0
df["Excel"] = df["Job Description"].apply(excel)
def hadoop(skill):
    if "hadoop" in skill.lower():
        return 1
    else:
        return 0
df["Hadoop"] = df["Job Description"].apply(hadoop)
def aws(skill):
    if "aws" in skill.lower():
        return 1
    else:
        return 0
df["AWS"] = df['Job Description'].apply(aws)
def python(skill):
    if "python" in skill.lower():
        return 1
    else:
        return 0
df ["Python "] = df["Job Description"].apply(python)
def spark(skill):
    if "spark" in skill.lower():
        return 1
    else:
        return 0
df["Spark"] = df["Job Description"].apply(spark)
def sql(skill):
    if "sql" in skill.lower():
        return 1
df["SQL"] = df["Job Description"].apply(sql)
def visualization(skill):
    if "tableau" or "power bi" in skill.lower():
        return 1
    else:
        return 0
df["Visualization"] = df["Job Description"].apply(visualization)
```
